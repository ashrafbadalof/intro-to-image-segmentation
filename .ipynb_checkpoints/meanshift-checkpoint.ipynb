```python
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift
from sklearn.cluster import estimate_bandwidth
from skimage.io import imread
from skimage.transform import resize
```

```python
image_name = "dlf_3833.jpg"
image = imread(image_name)
```

```python
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
image.shape
```

```python
image = resize(image, (334, 500))
```

```python
image.shape
```

```python
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
reshaped_image = image.reshape(-1, 3)
```

```python
reshaped_image.shape
```

```python
bandwidth = estimate_bandwidth(reshaped_image, quantile = 0.1)
```

```python
bandwidth
```

```python
ms = MeanShift(bandwidth = 0.14, n_jobs = -1)
```

```python
ms.fit(reshaped_image)
```

```python
labels = ms.labels_
cluster_centers = ms.cluster_centers_
```

```python
segmented_image = cluster_centers[labels].reshape(image.shape)

plt.figure()
plt.imshow(segmented_image)
plt.axis("off")
plt.show()
```
