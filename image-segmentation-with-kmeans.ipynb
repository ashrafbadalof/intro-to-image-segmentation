```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.cluster import KMeans
```

```python
image = mpl.image.imread("cat.jpg")
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
X = image.reshape(-1, 3)
X.shape
```

```python
kmeans = KMeans(n_clusters = 10, n_init = 10)
```

```python
kmeans.fit(X)
```

```python
segmented_im = kmeans.cluster_centers_[kmeans.labels_]
segmented_im = segmented_im.reshape(image.shape)
```

```python
plt.imshow(segmented_im / 255)
```

```python
import cv2

cv2.imwrite("imag2.png", cv2.cvtColor(segmented_im.astype("uint8"), cv2.COLOR_BGR2RGB))
```
